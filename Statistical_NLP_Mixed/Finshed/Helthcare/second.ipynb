```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np
```

```python
df=pd.read_csv('healthcare_dataset.csv')
```

```python
df['Test Results'].value_counts()
```

```python
#Data Cleaning 
```

```python

df['Test Results']=df['Test Results'].replace({'Inconclusive':0,'Normal':1,'Abnormal':2})
df.head()
```

```python
df.isnull().sum()
```

```python
df['Medical Condition'].value_counts()
df['Medication'].value_counts()
```

```python
# Encodeing  All Input Columns 
```

```python
# Data Spliting 
x=df.drop(columns=['Test Results'])
y=df['Test Results']
```

```python
# Feture Selection
x=x.drop(['Name','Date of Admission','Doctor','Discharge Date','Hospital'],axis=1)
```

```python
# Splitting 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python

# Now Encodeing label Encoder Menually 
x_train['Gender']=x_train['Gender'].replace({'Male':0,'Female':1})
x_test['Gender']=x_test['Gender'].replace({'Male':0,'Female':1})

#Label Encoder Apply in sklearn with Blood Type
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x_train['Blood Type']=le.fit_transform(x_train['Blood Type'])
x_test['Blood Type']=le.transform(x_test['Blood Type'])

# label Encoding in Insurance Provider
le=LabelEncoder()
x_train['Insurance Provider']=le.fit_transform(x_train['Insurance Provider'])
x_test['Insurance Provider']=le.transform(x_test['Insurance Provider'])
```

```python
#Label Encoder Apply in sklearn with Blood Type
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x_train['Medical Condition']=le.fit_transform(x_train['Medical Condition'])
x_test['Medical Condition']=le.transform(x_test['Medical Condition'])
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Define the ordinal structure (from least to most urgent)
urgency_order = [['Elective', 'Urgent', 'Emergency']]

# Initialize the encoder with the defined order
encoder = OrdinalEncoder(categories=urgency_order,handle_unknown='use_encoded_value',unknown_value=-1)

# Fit the encoder ONLY on training data
encoder.fit(x_train[['Admission Type']])  # Double brackets for DataFrame input

# Transform both train and test data
x_train['Encoded_adtype'] = encoder.transform(x_train[['Admission Type']])
x_test['Encoded_adtype'] = encoder.transform(x_test[['Admission Type']])

# Drop the original column
x_train= x_train.drop('Admission Type', axis=1)
x_test= x_test.drop('Admission Type', axis=1)
```

```python
# Medication lablel Encoder 
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x_train['Medication']=le.fit_transform(x_train['Medication'])
x_test['Medication']=le.transform(x_test['Medication'])
```

# Data Scaleing 


```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# নিউমেরিক কলাম নির্বাচন
numeric_cols = ['Age', 'Billing Amount', 'Room Number']

# স্কেলার ইনিশিয়ালাইজ
scaler = StandardScaler()

# শুধু ট্রেন ডেটায় ফিট করুন
scaler.fit(x_train[numeric_cols])

# ট্রেন এবং টেস্ট ডেটা স্কেল করুন
x_train[numeric_cols] = scaler.transform(x_train[numeric_cols])
x_test[numeric_cols] = scaler.transform(x_test[numeric_cols])

# এখন x_train এবং x_test-এ মূল কলামই স্কেল্ড ভ্যালু ধারণ করছে
```

# Traget Column Encodeing

# Model Apply 

```python
from sklearn.linear_model import LogisticRegression
```

```python
# মডেল ইনিশিয়ালাইজ
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# প্রেডিকশন
y_pred = model.predict(x_test)

# একুরেসি
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.9}")

# কনফিউশন ম্যাট্রিক্স
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# ক্লাসিফিকেশন রিপোর্ট
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
